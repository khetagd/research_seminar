```python
indexes = pd.read_excel("data_akbilgic.xlsx")
```

```python
indexes.drop('TL BASED', axis=1, inplace=True)
```

Все индексы выражены в долларах, поэтому удалим столбец ISE в лирах из датафрейма, он при анализе, в целом, не понадобится.

```python
names = list(indexes.iloc[0]) #в первой строке датафрейма находятся названия столбцов, сохраним их в отдельный список и в следующей строке удалим их из датафрейма
indexes.drop(0, inplace = True)
indexes.columns = names #присваиваем эти названия столбцам датафрейма
```

```python
indexes
```

```python
dates = indexes.date
indexes.drop('date',axis=1, inplace=True)
indexes = indexes.astype(float) #конвертируем все значения в float, предварительно убрав из датафрейма столбец с datetime значениями
indexes['dates'] = dates #возвращаем даты в обновленный датафрейм
```

```python
#посчитаем корреляцию
corr = indexes.drop('dates', axis=1).dropna().corr()
corr
```

```python
#выведем тепловую карту, чтобы было более наглядно видно корреляцию значений

from matplotlib import pyplot as plt
f, ax = plt.subplots(figsize=(10, 8))

sns.heatmap(corr,
    cmap=sns.diverging_palette(220, 10, as_cmap=True),
    vmin=-1.0, vmax=1.0,
    square=True, ax=ax)

plt.show()
```

Заметим, что сильнее всего коррелируют EU и DAX, EU и FTSE, DAX и FTSE. Это, в целом, логично, так как все они — европейские индексы.

Построим графики индексов

```python
def to_float(x): #функция меняет тип данных с object на float
    try:
        x = float(x)
    except ValueError:
        x = 0
    except TypeError:
        x = 0
    return x
```

Заменим значения таблиц на относительное изменение(относительно первого года)

```python
indexes = pd.read_excel("data_akbilgic.xlsx")
columns = list(indexes.iloc[0])
columns[1] = 'ISE_turkish'
indexes.columns = columns
indexes.iloc[0] = ['2009-01-04 00:00:00', '100','100','100','100', '100','100','100','100','100'] #добавляем стартовый день
indexes = indexes.drop(columns=["ISE_turkish"])
columns = indexes.columns
indexes.date = indexes.date.apply(lambda x: dt.fromisoformat(str(x))) #меняем тип данных на datetime
for column in columns[1:]:
    indexes[column] = indexes[column].apply(to_float)
for i in range(1, len(indexes)):
    for column in columns[1:]:
        indexes[column].iloc[i] = indexes[column].iloc[i - 1]*(indexes[column].iloc[i] + 1) #преобразуем изменение дневное в относительное
fig = go.Figure()
for column in columns[1:]:
    fig.add_trace(go.Scatter(x = indexes['date'], y = indexes[column], name = column)) #добавляем в график динамику индекса
fig.show()
```

Графики можно разбить на 3 группы: "ISE" , "EM" и "BOVESPA", оставшиеся соответственно и предсказания для них будут похожими.
